```python
# import library yang dibutuhkan
import pandas as pd
import numpy as np

# membuat dataset

sisi_1 = [22, 15, 9, 5, 17, 22, 16, 6, 6, 20, 8, 30, 9, 15, 11, 17, 5, 16, 9, 26]
sisi_2 = [10, 5, 5, 13, 6, 12, 8, 5, 16, 17, 5, 22, 12, 10, 15, 3, 6, 13, 15, 5]

# membuat data frame agar dapat dilakukan analisis data
df = pd.DataFrame({'sisi1': sisi_1, 'sisi2': sisi_2})
df
```

```python
# Fungsi menghitung persegi
def luasPersegi(sisi):
    '''Menghitung luas persegi
    |parameter sisi berupa float
    |fungsi akan mengembalikan nilai luas persegi'''
    return sisi * sisi
```

```python
# fungsi menghitung luas jajar genjang
def luasJajarGenjang(alas, tinggi):
    '''Menghitung luas jajar Genjang
    |parameter alas dan tinggi berupa float
    |fungsi akan mengembalikan nilai luas jajar genjang'''
    return alas * tinggi
```

```python
# membuat list luas 
luas = []

for i in range(len(df)):
    if df['sisi1'][i] > 10 : # luas adalah luas persegi jika sisi1 > 10
        luas.append(luasPersegi(df['sisi1'][i]))
    elif df['sisi1'][i] < 10: # luas adalah luas jajar genjang jika sisi1 < 10
        luas.append(luasJajarGenjang(df['sisi1'][i],df['sisi2'][i]))
    else: # luas adalah NaN apabila sisi1 = 10
        luas.append(None)

# tambahkan kolom luas pada data frame dengan nilai list luas sebelumnya
df['luas'] = luas
```

```python
# mengambil indeks array apabila sisi1 > 10 atau sisi2 bernilai ganjil 
# menggunakan library numpy

filtered_values = np.where((df['sisi1']>10) | (df['sisi2'] % 2 == 1))

# menghitung standard deviasi luas dari sisi1 > 10 atau sisi2 bernilai ganjil 

std_luas = df.loc[filtered_values]['luas'].std()
std_luas
```
