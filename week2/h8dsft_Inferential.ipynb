```python
# change type data Date from object to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Membuat dataframe baru
df_new = df[['Date',
                'Location',
                'New Cases',
                'New Deaths',
                'New Recovered',
                'New Active Cases',
                'Total Cases',
                'Total Deaths',
                'Total Recovered',
                'Total Active Cases'
                ]].query(("Location == 'DKI Jakarta' & Date.dt.year > 2021"))
# drop location column
df_new.drop(columns='Location',inplace=True)

df_new.head()
```

```python
# reset index from 0
df_new = df_new[df_new["Date"] != None ].reset_index(drop=True)
```

```python
df_new.head()
```

### Lakukan hypothesis testing terhadap dataset covid 19

```python
df_month_1_to_4  = df_new[df_new['Date'].dt.month <= 4][['Date','New Cases']]
df_month_5_to_9 = df_new[df_new['Date'].dt.month > 4][['Date','New Cases']]


print('Average New Cases a day in month 1 to 4 2022 is {}'.format(np.round(df_month_1_to_4['New Cases'].mean())))
print('Average New Cases a day in month 5 to 9 2022 is {}'.format(np.round(df_month_5_to_9['New Cases'].mean())))
```

```python
# set date to be index
df_month_1_to_4.set_index('Date',inplace=True)
df_month_5_to_9.set_index('Date',inplace=True)
```

mean1 = Average New Cases a day in month 1 to 4 2022 is 3187.0\
mean2 = Average New Cases a day in month 5 to 9 2022 is 1194.0

**Hypothesis :\
H0 : mean1 >= mean2\
H1 : mean1 < mean2**

Analisis menggunakan two sample independent two tailed

```python
from scipy import stats
t_stat, p_val = stats.ttest_ind(df_month_1_to_4,df_month_5_to_9)
print('P-value:',p_val[0]/2) #the p-value isn't divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat[0])
```

alpha = 0.05\
p-value = 0.000000012230945500014422\
p-value < alpha\
`H0 rejected`

### Kesimpulan dari hasil analisa.


Dari testing hipotesis diatas **H0 rejected**, kesimpulan yang dapat diambil adalah bahwa, kasus baru covid-19 di DKI Jakarta `per hari` pada januari, februari, maret dan april ***tidak signifikan lebih banyak*** dari mei, juni, juli, agustus dan september tahun 2022.
