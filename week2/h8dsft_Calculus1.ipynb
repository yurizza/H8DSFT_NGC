### Carilah turunan dari fungsi y


```python
# import library
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

#### Turunan Symbolic

```python
# turunan pertama dari y
x = sy.symbols('x')
y = x**2 + 2*x + 1 
dx = y.diff(x)
dx
```

```python
# turunan pertama dari y
x = sy.symbols('x')
y = 4*x**3 - 3*x**2 + 2*x - 1 
dx = y.diff(x)
dx
```

#### Turunan Numerik

```python
# membuat array dari -10 hingga 10 dibagi menjadi 100 element
x = np.linspace(-10,10,100)
# membuat fungsi Y
Y = x**2 + 2*x + 1 
```

```python
# Mencari turunan dengan turunan Y dibagi turunan x
dx = np.diff(Y)/np.diff(x)
dx
```

```python
# mencari turunan dengan gradient
dx_2 = np.gradient(Y,x)
dx_2
```

```python
# membuat plot dari turunan dx
plt.grid()
plt.plot(x[1:],dx) # (x, y) dengan x dimulai dari index 1 
```

```python
# membuat plot dari turunan dx_2
plt.grid()
plt.plot(x,dx_2) # (x, y) dengan x dimulai dari index 0
```
