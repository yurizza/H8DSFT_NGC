#### Loading Data

```python
# import all libraries
import pandas as pd
```

```python
# Loading data
df = pd.read_csv('Dataset/covid_19_indonesia_time_series_all.csv')
```

```python
# tampilkan all columns in dataframe
pd.set_option("display.max.columns", None)
```

```python
df.head()
```

```python
df.info()
```

```python
# change type data Date from object to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# make new dataframe, the Location is DKI Jakarta and year >= 2021
df_new = df[['Date',
             'Location',
             'Total Cases', 
             'Total Deaths',
             'Total Recovered',
             'Total Active Cases'
             ]].query(("Location == 'DKI Jakarta' & Date.dt.year >= 2021"))
df_new
```

```python
# reset index from 0
df_new = df_new[df_new["Location"] != None ].reset_index(drop=True)
df_new.info()
```

```python
# remove column Location
df_new.drop(columns='Location',inplace=True)
df_new
```

```python
# set date to be index
df_new.set_index('Date',inplace=True)
df_new
```

#### Cari Measure of Variance dari data tersebut.


```python
df_new.describe().T
```

```python
# Create a Histogram of Total Cases

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt


cases = df_new['Total Cases']
cases.plot.hist(title='Total Cases Distribution', color='lightblue', bins=25)  
plt.axvline(cases.mean(), color='magenta', linestyle='dashed', linewidth=2)
plt.axvline(cases.median(), color='green', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Get Skewness and Kurtosis Value

%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats

numcols = ['Total Cases', 'Total Deaths','Total Recovered','Total Active Cases']
for col in numcols:
    print(df_new[col].name + ' skewness: ' + str(df_new[col].skew()))
    print(df_new[col].name + ' kurtosis: ' + str(df_new[col].kurt()))
    density = stats.gaussian_kde(df_new[col])
    n, x, _ = plt.hist(df_new[col], histtype='step', bins=25)  
    plt.plot(x, density(x)*6)
    plt.show()
    print('\n')
```

##### Range

```python
numcols = ['Total Cases', 'Total Deaths','Total Recovered','Total Active Cases']

for col in numcols:
    print(df_new[col].name + ' range: ' + str(df_new[col].max() - df_new[col].min()))
```

##### Percentile and Quartile

```python
from scipy import stats

print('Percentile dari score 900000 total cases : ',stats.percentileofscore(df_new['Total Cases'], 900000, 'weak'))
print('Percentile dari score 16000 total deaths : ',stats.percentileofscore(df_new['Total Deaths'], 16000, 'weak'))
print('Percentile dari score 900000 total recovered : ',stats.percentileofscore(df_new['Total Recovered'], 900000, 'weak'))
print('Percentile dari score 18000 total active cases : ',stats.percentileofscore(df_new['Total Active Cases'], 18000, 'weak'))
```
