##### Buatlah sebuah kode untuk mensimulasikan melempar koin sebanyak 50.000 kali dengan assign value random berupa 0 dan 1 untuk kepala atau ekor. Setiap kali koin dilempar, peluang untuk mendapatkan kepala atau ekor adalah 50%.

```python
# import library
import random
from matplotlib import pyplot as plt
```

```python
# fungsi menghitung kemunculan heads and tails dalam percobaan
def randomJumlahHeadsTails(percobaan):
    '''Menghitung jumlah kemunculan heads atau tails dalam percobaan melempar koin
    |parameter lempar merupakan integer
    |mengembalikan nilai jumlah heads and tails'''

    # coin[0] untuk menghitung jumlah kemunculan heads
    # coin[1] untuk menghitung jumlah kemunculan tails
    coin = [0,0]
    for i in range(0,percobaan): # Berulang sebanyak percobaan
        ran = random.randint(0,1) # memberikan nilai random, 0 maka heads dan 1 tails.
        coin[ran] += 1 # increment heads or tails
    return coin 
```

```python
a = randomJumlahHeadsTails(percobaan=50000)
print (a)

# Show a pie chart of the results
plt.figure(figsize=(5,5))
plt.pie(a, labels=['heads', 'tails'])
plt.legend() 
plt.show()
```
